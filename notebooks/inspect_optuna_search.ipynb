# F-OPTUNA — search-behavior sanity checks

Data Validity group of `spec/2026-07-21-optuna-hyperparameter-search/validation.md`.

These check that the **search machinery works** — that the searched dimensions
actually move the objective, that pruning is live, and that both loss branches
are reachable. The per-trial budget here is deliberately tiny, so **none of the
angular-error numbers below are accuracy results** and must not be quoted as
such (see the R2 note in Roadmap.md).

```python
import sys, warnings
from pathlib import Path

REPO = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(REPO / "src"))
sys.path.insert(0, str(REPO / "scripts"))
warnings.filterwarnings("ignore")

import optuna, yaml
optuna.logging.set_verbosity(optuna.logging.WARNING)
print("repo:", REPO)
```

## Study config — 12 trials on a tiny subset

```python
BUNDLE = REPO.parent / "eve_shared" / "EveDataset" / "bundle"
CROPS  = REPO.parent / "eve_shared" / "EveDataset" / "eve_out"
OUT    = REPO / "runs" / "hpo_notebook"

SEARCH_SPACE = {
    "dropout1":     {"type": "float", "low": 0.0, "high": 0.7},
    "dropout2":     {"type": "float", "low": 0.0, "high": 0.7},
    "weight_decay": {"type": "float", "low": 1.0e-6, "high": 1.0e-2, "log": True},
    "hidden_dim":   {"type": "categorical", "choices": [128, 256]},
    "loss":         {"type": "categorical", "choices": ["angular", "cosine"]},
}

cfg = {
    "data": {"bundle_dir": str(BUNDLE), "crops_root": str(CROPS), "batch_size": 8,
             "num_workers": 0, "split_source": {"seed": 42, "val_fraction": 0.2}},
    "model": {"pretrained": False, "lr": 1.0e-3},
    "trainer": {"max_epochs": 3, "limit_train_batches": 3, "limit_val_batches": 2,
                "accelerator": "cpu", "log_every_n_steps": 1,
                "enable_checkpointing": False, "enable_progress_bar": False,
                "enable_model_summary": False},
    "output": {"dir": str(OUT)},
    "logging": {"wandb": {"enabled": False}},
    "optuna": {"study_name": "eyenet-hpo-nb", "storage": None, "direction": "minimize",
               "objective_metric": "val/angular_error_deg", "n_trials": 12, "timeout": None,
               "sampler": {"name": "tpe", "seed": 42},
               "pruner": {"name": "median", "n_warmup_steps": 1, "n_startup_trials": 3},
               "search_space": SEARCH_SPACE},
}
cfg_path = REPO / "runs" / "hpo_notebook_cfg.yaml"
cfg_path.parent.mkdir(parents=True, exist_ok=True)
cfg_path.write_text(yaml.safe_dump(cfg))
print("config written:", cfg_path)
```

```python
import tune
study = tune.main(str(cfg_path))
print(f"{len(study.trials)} trials")
```

## Trial table

```python
import pandas as pd
rows = [{"trial": t.number, "state": t.state.name, "value": t.value, **t.params}
        for t in study.trials]
df = pd.DataFrame(rows)
df
```

## Check 1 — the objective actually moves (std > 0)

A flat objective across trials would mean the search space or the wiring is
inert: hyperparameters sampled but never reaching the model.

```python
vals = df.loc[df["state"] == "COMPLETE", "value"].dropna()
std = float(vals.std())
print(f"completed={len(vals)}  mean={vals.mean():.3f} deg  std={std:.4f} deg")
print("PASS" if std > 0 else "FAIL: objective is flat -- search may be inert")
```

## Check 2 — pruning is live

With the median pruner and `n_startup_trials=3`, at least one of 12 trials should
be pruned. Zero pruned would mean the callback is a silent no-op that trains every
trial to completion — the exact failure that makes a "cheap" search expensive.

```python
n_pruned = int((df["state"] == "PRUNED").sum())
print(f"pruned={n_pruned} / {len(df)}")
print("PASS" if n_pruned >= 1 else "NOTE: 0 pruned -- re-check with more trials/epochs")
```

## Check 3 — best_params inside the declared bounds

```python
best = study.best_params
ok = True
for dim in ("dropout1", "dropout2", "weight_decay"):
    lo, hi = SEARCH_SPACE[dim]["low"], SEARCH_SPACE[dim]["high"]
    inside = lo <= best[dim] <= hi
    ok &= inside
    print(f"{dim:13s} {best[dim]:.6g}  in [{lo}, {hi}] -> {inside}")
for dim in ("hidden_dim", "loss"):
    inside = best[dim] in SEARCH_SPACE[dim]["choices"]
    ok &= inside
    print(f"{dim:13s} {best[dim]}  in {SEARCH_SPACE[dim]['choices']} -> {inside}")
print("PASS" if ok else "FAIL: an out-of-range value leaked from suggest dispatch")
```

## Check 4 — both loss branches are reachable

If the categorical `loss` dimension silently collapsed to one value, the search
would never compare the two losses it exists to compare.

```python
counts = df["loss"].value_counts().to_dict()
print(counts)
print("PASS" if len(counts) == 2 else f"NOTE: only {list(counts)} sampled at this seed")
```

## Check 5 — the handoff actually works

`best_params.yaml` is meant to be pasted into a `configs/*.yaml` `model:` block
for R3's full run. Verify `scripts/train.py`'s `main` accepts it verbatim.

```python
best_file = yaml.safe_load((OUT / "best_params.yaml").read_text())
print(best_file)

train_cfg = {
    "data": cfg["data"],
    "model": {"pretrained": False, "lr": 1.0e-4, **best_file["best_params"]},
    "trainer": {"max_epochs": 1, "limit_train_batches": 2, "limit_val_batches": 1,
                "accelerator": "cpu", "log_every_n_steps": 1,
                "enable_progress_bar": False, "enable_model_summary": False},
    "output": {"dir": str(REPO / "runs" / "hpo_handoff_check")},
    "logging": {"wandb": {"enabled": False}},
}
handoff_path = REPO / "runs" / "hpo_handoff_cfg.yaml"
handoff_path.write_text(yaml.safe_dump(train_cfg))

import train
train.main(str(handoff_path))
print("PASS: best_params.yaml is consumable by scripts/train.py")
```

## Summary

The search machinery is exercised end to end: dimensions reach the model, the
objective varies, pruning fires, sampled values stay in bounds, both losses are
reachable, and the `best_params.yaml` handoff is consumable by the full-run path.

⚠️ The angular-error values above come from a 3-epoch / 3-batch budget and are
**not** model-quality results. Producing the tuned settings for R3's full-split
run means re-running `py scripts/tune.py --config configs/optuna.yaml` at the
shipped budget.
